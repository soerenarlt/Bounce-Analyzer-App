```python
#IMPORTS
import os
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as la
```

```python
#os.chdir('still/')
df_nogacc = pd.read_csv('LinearAcceleration.csv')
print(df_nogacc)
```

```python
df_gacc = pd.read_csv('Accelerometer.csv')
print(df_gacc)
```

```python
t = df_nogacc['Time (s)']
ax = list(df_nogacc['Linear Acceleration x (m/s^2)'])
ay = list(df_nogacc['Linear Acceleration y (m/s^2)'])
az = list(df_nogacc['Linear Acceleration z (m/s^2)'])
a = np.array(list(zip(ax,ay,az)))

ax2 = list(df_gacc['Acceleration x (m/s^2)'])
ay2 = list(df_gacc['Acceleration y (m/s^2)'])
az2 = list(df_gacc['Acceleration z (m/s^2)'])
aplusg = np.array(list(zip(ax2,ay2,az2)))
```

```python
datapoints = min(len(t),len(ax))-1
plt.plot(t[:datapoints], ax[:datapoints], color="blue")
plt.xlabel("time")
plt.title("Acceleration")
plt.show()

datapoints = min(len(t),len(ay))-1
plt.plot(t[:datapoints], ay[:datapoints], color="blue")
plt.xlabel("time")
plt.title("Acceleration")
plt.show()

datapoints = min(len(t),len(ax))-1
plt.plot(t[:datapoints], az[:datapoints], color="blue")
plt.xlabel("time")
plt.title("Acceleration")
plt.show()
```

```python
i = 200
print(a[i])
print(aplusg[i])
diff = aplusg[i]-a[i]
print(diff)
print(la.norm(diff))
```

```python
#isolate up/down component of acceleration
vec = diff/la.norm(diff)
accz = np.dot(vec,a[i])
print(accz)
```

```python
len(df_gacc)
```

```python
#make list of accz
accz = []
for ind in range(len(df_gacc)):
    diff = aplusg[ind] - a[ind]
    vec = diff/(la.norm(diff)+1e-16)
    accz.append(np.dot(vec,a[ind]))
```

```python
datapoints = min(len(t),len(accz))-1
plt.plot(t[:datapoints], accz[:datapoints], color="blue")
plt.legend(["accz"])
plt.xlabel("time")
plt.ylabel("acceleration")
plt.title("Acceleration")
plt.show()
```

```python
tdiff=[t[i+1]-t[i] for i in range(len(t)-1)]
```

```python
vz=[0]
z=[0]
for i in range(datapoints):
    vz.append(vz[i]+accz[i]*tdiff[i])
    z.append(z[i]+vz[i]*tdiff[i])
```

```python
plt.plot(t[:datapoints], vz[:datapoints], color="blue")
plt.legend(["vz"])
plt.xlabel("time")
plt.ylabel("velocity")
plt.title("velocity")
plt.show()
```

```python
plt.plot(t[:datapoints], z[:datapoints], color="blue")
plt.legend(["z"])
plt.xlabel("time")
plt.ylabel("height")
plt.title("height")
plt.show()
```

```python
gvec=[]
for i in range(len(t)):
    diff = aplusg[i]-a[i]
    gvec.append(la.norm(diff))
```

```python
plt.plot(t, gvec, color="blue")
plt.xlabel("time")
plt.title("g")
plt.show()

plt.figure(
    figsize=(15, 6))
plt.plot(t, accz/max(accz), color="red")
plt.plot(t[:-1], tdiff/max(tdiff), color="blue")
plt.xlabel("time")
plt.title("tdiff")
plt.show()
```
