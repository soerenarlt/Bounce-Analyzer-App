```python
import os
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as la
```

```python
df_nogacc = pd.read_csv('LinearAcceleration.csv')
print(df_nogacc)
```

```python
df_gacc = pd.read_csv('Accelerometer.csv')
print(df_gacc)
```

```python
t = df_nogacc['Time (s)']
ax = list(df_nogacc['Linear Acceleration x (m/s^2)'])
ay = list(df_nogacc['Linear Acceleration y (m/s^2)'])
az = list(df_nogacc['Linear Acceleration z (m/s^2)'])
a = np.array(list(zip(ax,ay,az)))

ax2 = list(df_gacc['Acceleration x (m/s^2)'])
ay2 = list(df_gacc['Acceleration y (m/s^2)'])
az2 = list(df_gacc['Acceleration z (m/s^2)'])
aplusg = np.array(list(zip(ax2,ay2,az2)))
```

```python
df_gyro = pd.read_csv('Gyroscope.csv')
print(df_gyro)
```

```python
ox = list(df_gyro['Gyroscope x (rad/s)'])
oy = list(df_gyro['Gyroscope y (rad/s)'])
oz = list(df_gyro['Gyroscope z (rad/s)'])
o = np.array(list(zip(ox,oy,oz)))
```

```python
datapoints = min(len(t),len(ax))-1
plt.plot(t[:datapoints], ax[:datapoints], color="blue")
plt.xlabel("time")
plt.title("Acceleration")
plt.show()

datapoints = min(len(t),len(ay))-1
plt.plot(t[:datapoints], ay[:datapoints], color="blue")
plt.xlabel("time")
plt.title("Acceleration")
plt.show()

datapoints = min(len(t),len(ax))-1
plt.plot(t[:datapoints], az[:datapoints], color="blue")
plt.xlabel("time")
plt.title("Acceleration")
plt.show()
```

```python
#make list of accz
accz = []
for ind in range(len(df_gacc)):
    diff = aplusg[ind] - a[ind]
    vec = diff/(la.norm(diff)+1e-16)
    accz.append(np.dot(vec,a[ind]))
```

```python
datapoints = min(len(t),len(accz))-1
plt.plot(t[:datapoints], accz[:datapoints], color="blue")
plt.legend(["accz"])
plt.xlabel("time")
plt.ylabel("acceleration")
plt.title("Acceleration")
plt.show()
```
